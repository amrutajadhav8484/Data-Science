```python
#Data Integration
list1=[7058,7059,7075,7076]
list2=[7058,7059,7075,7076,7012]
li1=['A','B','C','D']
li2=['A','B','C','D','E']
df1=pd.DataFrame({'student ID':list1,'student name':li1})
print("first dataframe")
display(df1)
df2=pd.DataFrame({'student ID':list2,'student name':li2})
print("second dataframe")
display(df2)
merged=df2.merge(df1,how='left')
display(merged)
```

```python
#Data Integration
list1=[7058,7059,7075,7076]
list2=[7058,7059,7075,7076,7012,8756,8763,2345,9876]
li1=['A','B','C','D']
li2=['A','B','C','D','E','P','Q','R','S']
df1=pd.DataFrame({'student ID':list1,'student name':li1})
print("first dataframe")
display(df1)
df2=pd.DataFrame({'student ID':list2,'student name':li2})
print("second dataframe")
display(df2)
merged=df2.merge(df1,how='inner')
display(merged)
```

```python
#Data Integration
list1=[7058,7059,7075,7076]
list2=[7058,7059,7075,7076,7012]
li1=['A','B','C','D']
li2=['A','B','C','D','E']
df1=pd.DataFrame({'student ID':list1,'student name':li1})
print("first dataframe")
display(df1)
df2=pd.DataFrame({'student ID':list2,'student name':li2})
print("second dataframe")
display(df2)
merged=df2.concat(axis=1)
display(merged)
```
